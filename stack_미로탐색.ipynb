지나온 경로를 저장하는 방식으로 진행

탐색 알고리즘
1. 시작위치를 스택에 추가.
(현재 스택에는 시작위치만 있음)
2. 스택이 공백이 아니면, 하나의 위치를 꺼냄
(현재위치). 현재 위치에 “방문했음“ 표시.
만약, 스택이 공백이면, 출구 없음(종료)
3. 현재위치가 출구이면, 탐색 성공
4. 현재위치가 출구가 아니라면,
주변(상화좌우 or 좌우상하 등)을 탐색.
만약 주변이 아직 방문전이라면, 그 방의
위치를 모두 스택에 삽입
5. 단계 2로 돌아감

```python
class sta():
  def __init__(self):
    self.stack = []
  
  def isEmpty(self):
    """스택이 비어있으면 T, 아니면 F를 반환함"""
    if len(self.stack) == 0 :
      return True
    else :
      return False
  
  def push(self, e):
    self.stack.append(e)
    return self.stack
  
  def pop(self):
    """스택의 맨 위에 있는 항목을 꺼내서 반환"""
    a = self.stack.pop(-1)
    return a
  
  def peek(self):
    """스택의 맨 위에 있는 항목을 삭제하지 않고 반환"""
    if len(self.stack) > 0 :
      x = self.stack[-1]
      return x
    else : 
      print("항목을 꺼낼 수 없습니다. 확인요망")

  def size(self):
    if len(self.stack) > 0 :
      print("스택 내 모든 항목의 개수: ", len(self.stack))
    else :
      print("스택 내 항목이 없습니다.")

  def clear(self):
    self.stack.clear()
    return self.stack
  
  def findRoute(self, map, size):
  #입구 찾기
    for y in range(size):
      for x in range(size):
        a = map[y][x]
        if a == 'e' :
          self.push((y,x))
          a = '*'
          print('미로 입구: ', self.stack[0])
          print()

    while not self.isEmpty():
      #경로 찾기
      route = self.pop()                                                          # 1. 들어온 경로 값을 좌표값 형태(튜플)로 반환
      y = route[0]                                                                # 2. 튜플 값을 가져옴 
      x = route[1]
      map[y][x] = '5'                                                             # 3. 탈출경로

      scale = [(y-1,x),(y+1,x),(y,x+1),(y,x-1)]

      if route != None:
        for i in range(len(scale)):
          my = scale[i][0]
          mx = scale[i][1]

          if map[my][mx] != None : 
            if map[my][mx] == '0':
              self.push((my,mx))
              map[my][mx] = '*'
              print('(y,x)좌표:', my,mx)
              print('route',route)
              print('stack',self.stack)     

            elif map[my][mx] == 'x' :
              map[my][mx] = '**'
              self.push((my,mx))
              print('미로출구: ', '(',my,',',mx,')' )
              print('(y,x)좌표:', my,mx)
              print('route',route)
              print('stack',self.stack)   
              self.clear()
              break
            
            # elif map[my][mx] == '5' :
            #   map[my][mx] = '*'
            #   self.push(route)
            #   continue
              
            else : print('(y,x)좌표:', my,mx); print('route',route); print('stack',self.stack); continue
          else: break
       
      for k in map:
        print(k)
      print('========================')
                   
```
