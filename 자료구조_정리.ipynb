## 2. 집합

```python
class ADT() :#괄호를 해야한다...
  def __init__(self):
    self.set = []
  
  def size(self):
    """ 집합의 원소 개수 반환 함수"""
    x = len(self.set)
    return x
    
  def contains(self, e):
    """집합이 원소 e를 포함하는지를 검사하고 반환함"""
    if e in self.set :
      #print("{}는 집합에 포함되어 있습니다.").format(e)
      return self.set
    else :
      #print("{}는 집합에 포함되어 있지 않습니다.").format(e)
      print(self.set)
 
  def insert(self, e):
    """새로운 원소 e를 삽입, 이미 있다면 삽입하지 않음"""
    if e in self.set:
      #print("{}는 집합에 포함되어있습니다.").format(e)
      return self.set
    else :
      #print("{}는 집합에 없는 item입니다.").format(e)
      self.set.append(e)
      return self.set
    
  def delete(self, e):
    """원소 e를 집합에서 꺼내고 반환"""
    if e in self.set:
      self.set.remove(e)
      #print("{}를 집합에서 삭제하였습니다.").format(e)
      return self.set
    else :
      print("집합에 해당 item이 없습니다. 종료합니다.")
      
  def equals(self, setB):
    """setB와 같은 집합인지 검사, 더 좋은 방법이 있을듯(+예외)"""
    if self.set == setB :
      print("같은 집합 입니다.")
      return self.set, setB
    else :
      print("같은 집합이 아닙니다.")
      return self.set, setB

  def union(self, setB):
    """setB와 합집합을 만들어서 반환"""
    temp = []
    for i in self.set :
      temp.append(i)
    for j in setB :
      temp.append(j)
    return temp

  def intersect(self, setB):
    """setB와 교집합을 만들어 반환"""
    temp = []
    for i in self.set :
      for e in setB :
        if e == i :
          temp.append(e)
          return temp

  def difference(self, setB):
    """setB와 차집합을 만들어 반환"""
    for i in self.set:
      for e in setB :
        if i == e :
          self.set.remove(e)
      return self.set 
    
  def display(self):
    print(self.set)
    print("집합을 출력하였습니다.")
```

## 3.  스택

```python
class sta():
  def __init__(self):
    self.stack = []
  
  def isEmpty(self):
    """스택이 비어있으면 T, 아니면 F를 반환함"""
    if len(self.stack) == 0 :
      return True
    else :
      return False
  
  def push(self, e):
    self.stack.append(e)
    return self.stack
  
  def pop(self):
    """스택의 맨 위에 있는 항목을 꺼내서 반환"""
    a = self.stack.pop(-1)
    return a
  
  def peek(self):
    """스택의 맨 위에 있는 항목을 삭제하지 않고 반환"""
    if len(self.stack) > 0 :
      x = self.stack[-1]
      return x
    else : 
      print("항목을 꺼낼 수 없습니다. 확인요망")

  def size(self):
    if len(self.stack) > 0 :
      print("스택 내 모든 항목의 개수: ", len(self.stack))
    else :
      print("스택 내 항목이 없습니다.")

  def clear(self):
    self.stack.clear()
    return self.stack                   
```

```python
class Stack :
    def __init__( self ):   
        self.top = []       

    def isEmpty( self ): return len(self.top) == 0
    def size( self ): return len(self.top)
    def clear( self ): self.top = []	

    def push( self, item ):
        self.top.append(item)

    def pop( self ):
        if not self.isEmpty():
            return self.top.pop(-1)

    def peek( self ):
        if not self.isEmpty():
            return self.top[-1]

    def __str__(self ):
        return str(self.top[::-1])
```

```python
#괄호검사 교수님 예시 ==> stack 활용해서 다시 구조짜보기

def checkBrackets(statement):
    stack = Stack()
    for ch in statement:		    
        if ch in '{([': #if ch in ('{', '[', '('):	
            stack.push(ch)
        elif ch in ('}', ']', ')'):	
            if stack.isEmpty():
                return (False, "괄호의 시작이 부족하거나, 괄호의 끝이 더 많습니다.")
            else :
                left = stack.pop()
                if (ch == "}" and left != "{") or \
                   (ch == "]" and left != "[") or \
                   (ch == ")" and left != "(") :
                   return (False, "{}에 대응하는 괄호가 없습니다".format(ch))                

    #return stack.isEmpty()
    if stack.isEmpty():
        return (True, "모든 괄호가 적절하게 대응합니다")
    else:
        return (False, "대응되지 못한 괄호가 남아 있습니다")

str = ( "{ A[(i+1)] = 0; }", "if( (i==0) && (j==0 )", "A[ (i+1] ) = 0;" )
for s in str:
    m = checkBrackets(s)
    print(s," : ", m)
```

```python
# Queue ADT(원형큐)

max_qsize = 10
class CircularQueue:
  def __init__(self):
    self.front = 0
    self.rear = 0
    self.items = [None]*max_qsize

  def isEmpty(self):
    return self.front == self.rear

  def isFull(self):
    return self.front == (self.rear+1) % max_qsize
  
  def clear(self):
    '''공백상태로 만듦'''
    self.front = self.rear
  
  def enqueue(self, item):
    '''큐의 맨 뒤에 x를 추가함'''
    if not self.isFull():
       self.rear = (self.rear+1) % max_qsize
       self.items[self.rear] = item
  
  def dequeue(self):
    '''큐의 맨 앞 항목을 꺼내서 반환함'''
    if not self.isEmpty():
      self.front = (self.front+1) % max_qsize
      return self.items[self.front] 
  
  def peek(self):
    '''큐 앞에 있는 항목을 삭제하지 않고 반환'''
    if not self.isEmpty():
      return self.items[(self.front+1) % max_qsize]
  
  def size(self):
    '''큐의 모든 항목 개수를 반환'''
    if not self.isEmpty():
      return (self.rear - self.front + max_qsize) % max_qsize
```

```python
# 우선순위 큐 ADT
class PriorityQueue:
  def __init__(self):
    '''우선순위 큐 생성'''
    self.items = []

  def isEmpty(self):
    return len(self.items) == 0
  
  def enqueue(self, e):
    '''우선순위를 가진 항목 e를 추가'''
    self.items.append(e)

  def findMaxIndex(self):
    maxIndex = 0
    if not self.isEmpty():
      for i in range(1, self.size()):
        if self.items[i] > self.items[maxIndex] :
          maxIndex = i
          return maxIndex
    
  def dequeue(self):
    '''가장 우선순위가 높은 항목을 꺼내서 반환'''
    if not self.isEmpty():
      highest = self.findMaxIndex()
      return self.items.pop(highest)
  
  def peek(self):
    '''가장 우선순위가 높은 항목을 삭제하지않고 반환'''
    if not self.isEmpty():
      highest = self.findMaxIndex()
      return self.items[highest]

  def size(self):
    return self.items.size()
  
  def clear(self):
    return self.items.clear()
```
