exp(x) : 자연상수 e(2.7182818...)의 x 제곱을 반환함

expm1(x): 자연상수 e(2.7182818...)의 x 제곱에서 -1한 값 반환

==> pow나 다른 함수보다 자연상수 계산이 더 정확함!

```python
# exp(), expm1() 함수 활용
import math

print('e의 10제곱 값, ', math.exp(10))
print('e의 10제곱 -1 값, ', math.expm1(10))

#exp, expm1의 계산값과 차이가 있음을 확인
print('e ** 10 = ', (math.e ** 10))
print('pow(math.e,10) = ', pow(math.e, 10))
print('(e ** 10) -1 = ', (math.e ** 10)-1)
```

log(x): 진수가 x인 자연로그를 float값으로 반환함

log1p(x): 밑이 자연상수 e인 x+1의 자연로그를 float값으로 반환함

log2(x): 밑이 2인 x의 로그를 float값으로 반환함

log10(x): 밑이 10인 x의 로그를 float값으로 반환함

```python
# log(), log1p(), log2(), log10() 함수 활용
import math

a = math.log(10)
b = math.log(math.e)
print('log(10) = ', a)
print('log(math.e) = ', b ,'\n')

c = math.log1p(2)
d = math.log1p(math.e)
print('math.log1p(2) = ', c)
print('math.log1p(math.e) = ', d ,'\n')

e = math.log2(2)
f = math.log2(8)
print('math.log2(2) = ', e)
print('math.log2(8) = ', f ,'\n')

g = math.log10(10)
h = math.log10(10**2)
print('math.log10(10) = ', g)
print('math.log10(10**2) = ', h)
```
