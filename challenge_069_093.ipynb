# 대소문자 변경 함수

upper() : 모든 문자열을 대문자로 변경
lower() : 모든 문자열을 소문자로 변경


```python
#083 대소문자 구분

msg = input('대문자로 메세지를 입력하세요: ')

while True:
  if msg.isupper() == True :
    print(msg)
    break
  else :
    msg = input('대문자로 다시 입력하세요: ')
```

```python
#086 비밀번호 입력 + 일치여부
# ==> 문자는 일치하는데 대소문자가 틀린경우 (비밀번호를 lower로 한 뒤 맞는지 확인해보면 다른걸 알수있음)

password = input('비밀번호를 입력해주세요: ')
password2 = input('비밀번호를 한 번 더 입력해주세요: ')

check_up = password.lower()
check_up2 =password2.lower()

if password == password2 :
  print("Thank you")

elif check_up == check_up2 :
  print("They must be in the same case.")

else:
  print("Incorrect.")

```

```python
#092 배열 연결 expend()

arr = []
arr1 = []

for j in range(0,3):
  user = int(input('숫자를 입력하세요: '))
  arr.append(user)
print(arr)

for i in range(0, 5):
  num = random.randint(0, 100)
  arr1.append(num)
print(arr1)

if len(arr) > len(arr1):
  arr.expend(arr1)
  arr.sort
  print('arr: ', arr)
  for i in arr :
    print(i)

elif len(arr) == len(arr1) :
  print("두 배열의 길이가 같습니다.")

else:
  arr1.extend(arr)
  arr1.sort()
  print('arr1: ', arr1)
  for i in arr1:  
    print(i)
```
