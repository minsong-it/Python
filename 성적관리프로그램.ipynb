문제 : 국어 성적 관리 프로그램

<조건>
- 학년은 6학년까지
- 학년당 반은 5 개
- 한반에 학생 수  5, 4, 5, 3, 4명 (총 21)

1. 국어 점수 입력받음, 저장함. 단, 점수의 범위는 0~100점 사이
2. 어떤 학생들은 점수가 음수이거나 100점 이상으로 입력 각 학년에 모두 이런 실수가 있음
3. 실수를 바로 잡고 싶음
4. 그런데 점수를 전부 다 다시 입력받는 작업은 하고 싶지 않음
5. 잘못 입력한 점수를 찾아내고 그 부분만 새롭게 작업
6. 이런 작업을 함수를 사용해 작성
<작성 함수>
- 학생들의 국어점수 입력 함수 grade()
- 점수를 검증하는 함수(잘못된 입력을 찾는 함수)
- 점수를 수정할 수 있는 함수
7. 각 학년에서 평균점수가 가장 높은 반과 가장 낮은 반을 찾아 교장선생님께 알려드려야 함
8. 학년의 평균 점수보다 낮은 점수를 가진 학생들은 방과 후에 보충 수업을 해야 함, 그 대상이 누구인지 찾아야 함(학생명단 작성)

(랜덤함수 사용해서 대입하는 방법도 고민해볼것)

함수 목록


---



*   성적 입력: score_class()
*   성적 검증 + 수정: score_check()
*   평균 점수 계산



```python
def score_class(L):
  '''학생 성적 입력, class 리스트 생성 및 성적 추가'''
  for i in range (1,6) :
    score = list(map(int,input('학생 성적을 입력해주세요>> ').split()))
    L.append(score)
    print('{0}반 : {1}'.format(i, score))
  return L
```

```python
'''
애들 성적
1반 100 70 35 80 95
2반 90 140 35 70 
3반 55 85 90 -20 70
4반 45 100 200
5반 -50 80 95 100
'''
```

```python
grade1 = []
score_class(grade1)     #1학년 반 성적 추가
```

```python
def score_check(L):
  '''성적 오류 체크'''
  for i in range(len(L)):
    for j in range(len(L[i])):                                                    #range(len(L))로 진행할 경우 outOfIndex 에러 발생, 앞서 불러온 i만큼 길이 반복으로 줄여줄 것
      if L[i][j] > 100 or L[i][j] < 0 :                                           #100점 이상, 음수인 점수 체크
        print('{0}반의 {1}번 학생의 {2}점은 수정이 필요합니다'.format(i,j,L[i][j])) #수정 대상 출력
        re_score = input('수정 성적 입력: ')
        L[i][j] = int(re_score)                                                  #input으로 받은 값을 int로 변환해서 대입 => map(int(input())으로 받아서 변수를 바로 대입하는 경우 주소값을 반환하기 때문
        # L.insert((i,j),rescore)                                                #이중배열에서 insert는 사용불가, 인덱스부분에 튜플이 들어가지 않음
  return L
```

```python
score_check(grade1)     #성적오류 체크
```

```python
def class_mean(L):
  '''성적 평균 계산'''
  class_mean_score = []
  count = 0
  #반 성적 평균
  for i in L:
    class_sum = sum(i)
    class_mean = class_sum / len(L)
    class_mean_score.append(class_mean)
  return class_mean_score
```

```python
def avg_class(L):
  # 평균이 높은 반, 낮은 반 반환
  cm = class_mean(L)
  min_class = min(cm)
  max_class = max(cm)
  for i in cm:
    if i == min_class : print('평균이 가장 높은 반: {0} \n평균점수: {1}'.format(cm.index(i)+1,i))
    elif i == max_class : print('평균이 가장 낮은 반: {0} \n평균점수: {1}'.format(cm.index(i)+1,i))
```

```python
def grade_mean(L):
  #학년 성적평균
  grade_mean_score = 0
  cl = class_mean(L)
  grade_mean_score = sum(cl) / len(cl)
  return grade_mean_score
```

```python
cm = class_mean(grade1)
for i in range(len(cm)):
  print('{0}반 성적 평균: {1}'.format(i, cm[i]))
print()
avg_class(grade1)
print()
grade_mean(grade1)
```

```python
def student_check(L):
  '''학년 평균 점수보다 미달된 학생 탐색'''
  gm = grade_mean(L)
  print('<미달 학생 목록>')
  for i in range(len(L)):
    for j in range(len(L[i])):
      if L[i][j] < gm:
        print('{0}반 {1}번: 국어 {2}점'.format(i,j,L[i][j]))
```

```python
student_check(grade1)
```
