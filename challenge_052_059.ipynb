```python
#054
#임의출력-입력받음-두개 비교-임의값 출력
import random

C = ['h','r']
cho = random.choice(C)
use = input('어떤 것을 고르시겠습니까>> ')

if cho == use :
  print("You win")
else :
  print("Bad luck")

print('컴퓨터는',cho,'를 선택했습니다.')
```

```python
#057 random 활용
import random

num = random.randint(1,10)

while True :
  user = int(input('숫자 입력>> '))
  if num > user :
    print('숫자가 작습니다')
    user = int(input('다시 입력>> '))
  elif num < user :
    print('숫자가 큽니다')
    user = int(input('다시 입력>> ')) 
  else :
    print('같습니다.')
    break

print('컴퓨터가 출력한' ,num)
```

```python
#058 수학퀴즈 만들기
import random

score = []

for i in range(0,5) :
  x = random.randint(0,100)
  y = random.randint(0,100)
  print(x,'+',y,'= ?' )
  answer = int(input('정답>> '))
  
  if answer == x+y :
    score.append(i)
  else :
   print("틀렸습니다")

print('내 점수: ',len(score),'/5')
```

```python
#059 색상 맞추기
import random
color = ['빨','주','노','초','파']
R = random.choice(color)
re = input('빨,주,노,초,파 중에서 선택>> ')

while True :
  if R == re :
    print('컴퓨터가 선택한 것>> ',R, "\nWell done")
    break
  else :
    if R == color[0] :
      print("컴퓨터는 빨광색")
      break
    elif R == color[1]:
      print("컴퓨터는 주황색")
      break
    elif R == color[2]:
      print("컴퓨터는 노랑색")
      break
    elif R == color[3] :
      print("컴퓨터는 초록쌕")
      break
    elif R == color[4] :
      print("컴퓨터는 파랑쉑")
      break
```
