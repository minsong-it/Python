```python
class sta():
  def __init__(self):
    self.stack = []
  
  def isEmpty(self):
    """스택이 비어있으면 T, 아니면 F를 반환함"""
    if len(self.stack) == 0 :
      return True
    else :
      return False
  
  def push(self, e):
    self.stack.append(e)
    return self.stack
  
  def pop(self):
    """스택의 맨 위에 있는 항목을 꺼내서 반환"""
    return self.stack.pop(-1)
  
  def peek(self):
    """스택의 맨 위에 있는 항목을 삭제하지 않고 반환"""
    if len(self.stack) > 0 :
      x = self.stack[-1]
      return x
    else : 
      print("항목을 꺼낼 수 없습니다. 확인요망")

  def size(self):
    if len(self.stack) > 0 :
      print("스택 내 모든 항목의 개수: ", len(self.stack))
    else :
      print("스택 내 항목이 없습니다.")

  def clear(self):
    self.stack.clear()
    return self.stack

  def cal(self, L) :
    '''후위 수식계산'''
    code = ['/','*','-','+']               # 연산자를 만나면 스택에 넣어둔 숫자를 계산
    if self.isEmpty() != None :
      for i in L:
        if i in code :
          a = float(self.pop())
          b = float(self.pop())
          if i == '/':
            self.push(a / b)
          elif i == '*':
            self.push(a*b)
          elif i == '-':
            self.push(a-b)
          elif i == '+':
            self.push(a+b)
        else:
          self.push(i)                    # 숫자일 경우 스택에 추가함
    return self.pop()
  
  def postfix(self,L):
    '''중위 -> 후위 변환'''
    if len(L) != 0 :
      code = ['/','*','-','+']  
      for i in L:
        if i in code :
          self.push(i)
          x = self.pop()  
          if i == '-''+' and x == '/''*' :
            self.push(i)  
            print(self.stack)
        else :
          print(i)
      # print(self.stack)
```

```python
s = sta()
c = ['8','2','/','3','-','3','2','*','+']
s.postfix(c)
print('후위수식 계산: ', s.cal(c))
```
