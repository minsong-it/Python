(List) sort(): 리스트 item을 오름차순으로 정렬
==> 정렬을 위해선 int,str로 타입이 같아야 함!

==> sort()를 사용하면 리스트가 정렬되기 때문에 식별자에 대입하고 싶다면 .copy() 사용

```python
#List sort() 함수 활용

i = [1,2,3,4,5,9,8,7]
#j = ['아','이','우','에',5]

i.sort()

a = i.copy()
b = i
print(i)
print(a)
print(b)

# 저장된 메모리 주소 비교
print(id(a))
print(id(b))
print(id(i))

# b = j.sort(), print(b) 
# 에러타입- TypeError: '<' not supported between instances of 'int' and 'str'
```

(Dictionary) clear(): 딕셔너리의 모든 item을 삭제함

==> 빈 딕셔너리로 초기화한다면?  객체 주소가 남아있음

==>clear() 참조된 모든 객체를 삭제함

```python
#Dictionary {} 초기화

dic = {'a':1, 'b':2, 'c':3 }
dic2 = dic

print(dic)
print(dic2)
print(id(dic))

dic = {}

print(dic)  #빈 딕셔너리 객체 새로 생성
print(dic2) #기존 객체 유지

print(id(dic))
print(id(dic2)) #기존 객체주소와 동일
```

```python
#Dictionary clear() 함수 활용

dic = {'a':1, 'b':2, 'c':3 }
dic2 = dic

dic.clear()  #참조된 모든 객체 삭제

print(dic)
print(dic2)
```

(Dictionary) copy(): 딕셔너리를 다른 객체로 복사함

==>복사한 딕셔너리는 다른 객체로 저장됨!!

```python
#Dictionary copy() 함수 활용

x = {'a':0, 'b':1, 'c':2}
y = x

#is 연산자로 같은 딕셔너리인지 확인
print(x is y)

y = x.copy()
print(x is y)

print(id(x))
print(id(y))
```

(Dictionary) fromkeys(seq[, v]) 딕셔너리의 key값(+value) 을 반환함

```python
#Dictionary fromkeys() 함수 활용

keys = {'s','e','o','u','l'}

cities = dict.fromkeys(keys)
print(cities)

#value값 추가하기
value = [1] 
cities = dict.fromkeys(keys, value)
print(cities)

value.append(10)
print(cities)
```

(Dictionary) setdefault(key,val):  해당 key값이 있다면 value를 반환하고, value값이 없다면 val값을 저장한 뒤 반환함.

```python
#Dictionary setdefault() 함수 활용

x = 'Hello!'
y = {}

for i in x :
  y[i] = y.setdefault(i, 0)

print(y)
```

(Dictionary) update([other]): 딕셔너리의 데이터를 여러 개 수정할 수 있음, 딕셔너리 객체나 키/값 쌍으로 구성된 데이터를 받아 딕셔너리를 갱신함.

==> 공통된 key값을 가진 딕셔너리를 update하는 경우, 덮어쓰기한 값으로 변환됨


```python
#Dictionary update() 함수 활용

dict1 = {'A': 1, 'B': 2, 'C': 3}
dict2 = {'C': 2, 'D': 4, 'E': 5}

dict2.update({'F': 6, 'G': 7 })
print(dict2)

dict1.update(dict2)
print('병합된 딕셔너리>> ', dict1)

# 'C'값이 덮어쓰기한 dict2의 값 2임을 확인
```

```python
# Dictionary 딕셔너리끼리 병합

dict1 = {'A': 1, 'B': 2, 'C': 3}
dict2 = {'C': 2, 'D': 4, 'E': 5}

# dict(대상 딕셔너리, **덮어쓰는 딕셔너리)
print('C값이 2인 딕셔너리로 병합')
print(dict(dict1, **dict2))
print('C값이 3인 딕셔너리로 병합')
print(dict(dict2, **dict1))
```

(literal) 내장함수

** *iterable의 의미는 element를  하나씩 차례로 반환 가능한 object를 말하며
list, tuple, dictionary이 해당됨***

all(): 모든 요소가 참인가? T/F

 any(): 요소 중 어느 하나라도 참인가? T/F

 len(): 객체의 길이 수 반환

 cmp(): 두 iterable 의 항목을 비교합니다. (Python 3에서는 사용할 수 없음)

 sorted(): key값 오름차순으로 정렬


```python
menu = {'짜장면': 0, "짬뽕": 1, "볶음밥":2, "야끼우동":3, "중화비빔밥":4 }

print('all()>> ', all(menu))
print('any()>> ', any(menu))
print('len()>> ', len(menu))
print('sorted()>> ', sorted(menu))
```
