#힙 구현 : 리스트 사용

```python
# 6/27 class 코드 리뷰

class MaxHeap:
  def __init__(self):
    self.heap = []
    self.heap[0] = 0                  #힙에서 인덱스 0은 사용하지 않기 때문
  
  def size(self): return len(self)- 1
  
  def isEmpty(self): return self.size() == 0

  def Parent(self, ind):
    '''자식 인덱스를 넣었을 때 부모의 값을 반환'''
    return self.heap[ind/2]
  
  def Left(self, ind):
    '''해당 index 값의 왼쪽 자식 반환'''
    return self.heap[ind*2]
    # if self[index].left != None :
    #   return self[index].left ==> 트리 구조가 아니기때문에 왼쪽자식 인덱스를 구해야 함

  def Right(self, ind):
    '''해당 index 값의 오른쪽 자식 반환'''
    return self.heap[ind*2+1]
```
