# 특수메서드


---


> 클래스 내부에 선언할 수 있는 특수 메서드
> 파이썬 기존의 연산자, 함수를 오버 로딩하기 때문에 정해진 이름과 형식을 따라야 함

*   __ 함수 명 __() : 메서드 앞 뒤로 언더바
*   메서드 호출은 통용되는 연산자, 함수명으로 호출함 (__add__()의 경우:  + , add()로 호출가능)






```python
class Point():      
  #class명은 ()를 붙이지 않아도 가능함
  
  def __init__(self, x=0, y=0):
    ''' 클래스 생성자(클래스 객체) 생성'''
    self.x = x
    self.y = y
    # x = self.x를 할 경우 x의 속성이 없기 때문에 attribute 오류발생 

  def __str__(self):
    '''출력 형식 지정'''
    return "({},{})".format(self.x, self.y)

    #format의 경우 {0},{0}으로 인덱스 지정을 해놓으면 같은값이 계속 출력, 인덱스를 넘어가는 수이면 Index에러 발생

  def __add__(self, other):
    '''생성자 x끼리, y끼리 연산을 위해 self,other로 받음'''
    x = self.x + other.x
    y = self.y + other.y
    return Point(x, y)

  def __sub__(self,other):
    x = self.x - other.x
    y = self.y - other.y
    return Point(x, y)
  
  def __mul__(self, other):
    x = self.x * other.x
    y = self.y * other.y
    return Point(x, y)    
  
  def __pow__(self, other):
    x = self.x ** other.x
    y = self.y ** other.y
    return Point(x, y)

  def __truediv__(self, other):
    x = self.x / other.x
    y = self.y / other.y
    return Point(x, y)

  def __floordiv__(self, other):
    x = self.x // other.x
    y = self.y // other.y
    return Point(x, y)

  def __mod__(self,other):
    x = self.x % other.x
    y = self.y % other.y
    return Point(x, y)
```

```python
p1 = Point(1, 1)
p2 = Point(2, 3)

print("p1.__add__(p2) : ", p1.__add__(p2))
print("+ 연산자 사용: ", p1 + p2)
print('='*20)
print("p1.__sub__(p2) : ", p1.__sub__(p2))
print("- 연산자 사용: ", p1 - p2)
print('='*20)
print("p1.__mul__(p2) : ", p1.__mul__(p2))
print("* 연산자 사용: ", p1 * p2)
print('='*20)
print("p1.__pow__(p2) : ", p1.__pow__(p2))
print("** 연산자 사용: ", p1 ** p2)
print('='*20)
print("p1.__truediv__(p2) : ", p1.__truediv__(p2))
print("/ 연산자 사용: ", p1 / p2)
print('='*20)
print("p1.__floordiv__(p2) : ", p1.__floordiv__(p2))
print("// 연산자 사용: ", p1 // p2)
print('='*20)
print("p1.__mod__(p2) : ", p1.__mod__(p2))
print("% 연산자 사용: ", p1 % p2)
print('='*20)
```

```python
class Bit:
  '비트 연산자 계산 class'
  def __init__(self, x, y):
    self.x = x
    self.y = y
  
  def __str__(self):
    return "({},{})".format(self.x, self.y)
  
  def __lshift__(self, other):
    x = self.x << other.x
    y = self.y << other.y
    return Bit(x, y)                  

  def __rshift__(self, other):
    x = self.x >> other.x
    y = self.y >> other.y
    return Bit(x, y)

  def __and__(self, other):
    x = self.x & other.x
    y = self.y & other.y
    return Bit(x, y)                  

  def __or__(self, other):
    x = self.x | other.x
    y = self.y | other.y
    return Bit(x, y)                  
  
  def __xor__(self, other):
    x = self.x ^ other.x
    y = self.y ^ other.y
    return Bit(x, y)                  
  
  def __invert__(self):
    x = ~self.x
    y = ~self.y
    return Bit(x, y)                  
```

```python
p1 = Bit(1, 1)
p2 = Bit(2, 3)

print("p1.__lshift__(p2) : ", p1.__lshift__(p2))
print("<< 연산자 사용: ", p1 << p2)
print('='*20)
print("p1.__rshift__(p2) : ", p1.__rshift__(p2))
print(">> 연산자 사용: ", p1 >> p2)
print('='*20)
print("p1.__and__(p2) : ", p1.__and__(p2))
print("& 연산자 사용: ", p1 & p2)
print('='*20)
print("p1.__or__(p2) : ", p1.__or__(p2))
print("| 연산자 사용: ", p1 | p2)
print('='*20)
print("p1.__xor__(p2) : ", p1.__xor__(p2))
print("^ 연산자 사용: ", p1 ^ p2)
print('='*20)
print("p1__invert__() : ", p1.__invert__())
print("~ 연산자 사용: ", ~p1)
print('='*20)
```

```python
class Sequence:
  '''비교 연산자 계산 class'''
  def __init__(self, x, y):
    self.x = x
    self.y = y
  
  def __str__(self):
    return "({},{})".format(self.x, self.y)
  
  def __lt__(self, other):
    x = self.x < other.x
    y = self.y < other.y
    return Seqence(x, y)

  def __le__(self, other):
    x = self.x <= other.x
    y = self.y <= other.y
    return Seqence(x, y)

  def __eq__(self, other):
    x = self.x = other.x
    y = self.y = other.y
    return Seqence(x, y)

  def __ne__(self, other):
    x = self.x != other.x
    y = self.y != other.y
    return Seqence(x, y)

  def __gt__(self, other):
    x = self.x > other.x
    y = self.y > other.y
    return Seqence(x, y)

  def __ge__(self, other):
    x = self.x >= other.x
    y = self.y >= other.y
    return Seqence(x, y)
```

```python
p1 = Sequence(1, 1)
p2 = Sequence(2, 3)

print("p1.__lt__(p2) : ", p1.__lt__(p2))
print("< 연산자 사용: ", p1 < p2)
print('='*20)
print("p1.__le__(p2) : ", p1.__le__(p2))
print("<= 연산자 사용: ", p1 <= p2)
print('='*20)
print("p1.__eq__(p2) : ", p1.__eq__(p2))
print("== 연산자 사용: ", p1 == p2)
print('='*20)
print("p1.__ne__(p2) : ", p1.__ne__(p2))
print("!= 연산자 사용: ", p1 != p2)
print('='*20)
print("p1.__gt__(p2) : ", p1.__gt__(p2))
print("> 연산자 사용: ", p1 > p2)
print('='*20)
print("p1__ge__() : ", p1.__ge__(p2))
print(">= 연산자 사용: ", p1 >= p2)
print('='*20)
```
