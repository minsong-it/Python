```python
transposed = []
matrix = [[1,2,3,4],[4,5,6,8]]

for i in range(len(matrix[0])) :   #4개의 list 생성 0~3번째
  transposed_row = []

  for row in matrix :                #row는 matrix의 인자들로 구성, 반복
      transposed_row.append(row[i])   #[]에 하나씩 추가
  transposed.append(transposed_row)   #transposed = [] 에 _row 리스트를 넣음

print(transposed)
```

```python
n = int((input("Enter n: ")))

sum = 0
i = 1

while i <= n:   #1~입력받은 수까지의 합
  sum = sum + i
  i = i+1        #i++ 
print("The sum is",sum)
```

```python
odd = [1,3,5]
odd[0] = 0  #odd[0]값 대체
print(odd)  

odd[3:5] = [-5,-2] #odd[3] = -5, odd[4] = -2 ==> 슬라이싱을 통한 추가
print(odd)

odd.append(7)   #0,3,5,-5,-2,7
print(odd)   

odd.extend([9, 11, 13])  #맨 마지막 추가
print(odd)

odd.insert(0,10)  #odd[0]부터 추가
print(odd)

print(odd + [9,7,5]) # 연산자로도 추가 가능

print(["re"]*3)
```

```python
squares = {1:1, 2:4, 3:9, 4:16, 5:25}
print(squares.pop(4))  #키값인 4를 pop
print(squares)         #key, value가 다 없어짐
# print(squares.pop(9)) value 값은 pop x
print('popitem>>',squares.popitem())  #popitem은 LIFO 순서로 pop되면서 지정못함

del squares   #dict 제거
# print(squares)  NameError: name 'squares' is not defined
```

```python
squ = {1:1, 2:4, 3:9, 4:16, 5:25}

print(all(squ))   #모든 요소가 참이면 all, 0:0이 있으면 f
print(any(squ))   #어떤 한 요소라도 참이면 t
print(len(squ)) 
print(sorted(squ)) #key값을 sort
print(squ.items())   # ([(1, 1), (2, 4), (3, 9), (4, 16), (5, 25)]) items 출력
print(squ.setdefault(2))  #key 검색
```
